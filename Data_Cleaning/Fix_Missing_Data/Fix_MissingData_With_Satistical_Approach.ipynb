```python
import missingno as msno
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("NFL_clean_data.csv") #data src: https://www.kaggle.com/code/rtatman/data-cleaning-challenge-handling-missing-values/input?select=NFL+Play+by+Play+2009-2016+%28v3%29.csvdf
df
```

```python
msno.matrix(df)
```

```python
print(df.iloc[:50,:50])
msno.matrix(df.iloc[:,:50])
```

```python
df['down'] = df['down'].replace(np.NAN, df['down'].mean())
```

```python
print(df.iloc[:50,:50])
msno.matrix(df.iloc[:,:50])
```

```python
df['FirstDown'] = df['FirstDown'].replace(np.NAN, df['FirstDown'].median())
```

```python
print(df.iloc[:50,:50])
msno.matrix(df.iloc[:,:50])
```

Benefits of this method:  
Avoid data loss as compared to 'Remove row and column' Method.  
Can also useful with small datasets.

Drawbacks of this Method:  
Works only in numerical data. (if data is in non-numerical format then we have to do normalization first and then can use this method)  
Maybe be the reason of the data leakage.
