## "Backward fill" for Rows,

```python
#For df,

# 2    3    4    NaN
# 3    4    NaN  5 
# 1    NaN  3    2
# NaN  2    NaN  1

# Every NaN will be replaced by previous Row's value at the same column

# 2    3    4    5
# 3    4    3    5 
# 1    2    3    2
# NaN  2    NaN  1
```

## "Backward fill" for Columns,

```python
#For df,

# 2    3    4    NaN
# 3    4    NaN  5 
# 1    NaN  3    2
# NaN  2    NaN  1

# Every NaN will be replaced by previous Coulumn's value at the same Row

# 2    3   4    NaN
# 3    4   5    5 
# 1    3   3    2
# 2    2   1    1
```

```python
df.bfill(axis = 0) # " axis = 0 or axis = 'rows' " - means for Rows(Indexes)
```

```python
df.iloc[:30,:]['missing']
```

```python
df.isna().sum()
```

```python
df = pd.read_csv("MissingMigrantsProject.csv", encoding='latin-1')
df
```

```python
df = df.bfill(axis = 1) # " axis = 1 or axis = 'columns' " - means for Columns
```

```python
df.isna().sum()
```

```python
df = pd.read_csv("MissingMigrantsProject.csv", encoding='latin-1')
df
```

```python
df['missing'] = df['missing'].bfill(axis = 'rows')  # " axis = 0 or axis = 'rows' " - means for Rows(Indexes)
```

```python
df.iloc[:30,:]['missing']
```

```python
df.isna().sum()
```

```python
df.loc[0:30, ['affected_nationality','region_origin']]
```

```python
df.loc[:, ['affected_nationality','region_origin']] = df.loc[:, ['affected_nationality','region_origin']].bfill(axis = 'columns')  # " axis = 1 or axis = 'columns' " - means for Columns
```

```python
df.loc[0:30, ['affected_nationality','region_origin']]
```
