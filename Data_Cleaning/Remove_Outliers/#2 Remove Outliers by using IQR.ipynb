```python
# Quantile is a mathematical term

#quantile calculates the value below which a certain proportion of the data falls.
#Ex.

#if df is,
# id  value
# 0   10
# 1   20
# 2   30
# 3   40
# 4   50

# df.value.quantile(0.25) =  15.0
# df.value.quantile(0.5) =  30.0  (Note: quantile(0.5) is Median too.)
# df.value.quantile(0.75) =  45.0
```

```python
q1 = df.f_00.quantile(0.25)
q3 = df.f_00.quantile(0.75)
q1, q3
```

```python
Iqr = q3 - q1 #find Iqr
Iqr
```

```python
min = q1 - 1.5*Iqr #Calculate lower fence
max = q3 + 1.5*Iqr #Calculate upper fence
min, max
```

```python
#all outliers
df2 = df[(df.f_00<min)|(df.f_00>max)]
df2
```

```python
df = df[(df.f_00>min)&(df.f_00<max)]
df
```

```python
y = np.arange(1, df.shape[0]+1) 
plt.scatter(df["f_00"], y) 
plt.xlabel = "f_00"
plt.show()
```
