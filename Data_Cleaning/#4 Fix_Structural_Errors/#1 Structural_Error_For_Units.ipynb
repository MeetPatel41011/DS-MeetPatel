```python
import pandas as pd
```

```python
df = pd.read_csv("Student's_data.csv")
df #Two types of units are there "lb" and "kg", there should be only one unit in whole dataset for accurate statistics (Universal fact)
```

```python
df2 = df.copy()
df['Weight'] = df['Weight'].apply(lambda x: str(float(x.split()[0])*2.20462) + ' lb' if 'kg' in x else x)


# In this example code, we first read in the original CSV file using Pandas' read_csv() function. Then, we use the apply() function to apply a lambda function to each row in the 'Weight' column. This lambda function first checks if the weight value contains the string 'kg'. If it does, it converts the weight value to pounds by multiplying it by the conversion factor 2.20462, and then concatenates the string 'lb' to the end of the converted value. If the weight value does not contain the string 'kg', the lambda function simply returns the original value. Finally, we write the updated data to a new CSV file using the to_csv() function, setting the index parameter to False so that the row indices are not included in the output file.

# Note that in this example code, we assume that the original CSV file is named 'students.csv', and that it is located in the same directory as the Python script. The updated data is written to a new CSV file named 'students_updated.csv', also located in the same directory as the Python script. You can customize these file names and paths as needed to fit your own use case.
```
