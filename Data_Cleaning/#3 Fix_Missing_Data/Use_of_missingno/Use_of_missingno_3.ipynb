```python
import missingno as msno
import pandas as pd
```

```python
df = pd.read_csv("NFL Play by Play 2009-2016 (v3).csv") #data src: https://www.kaggle.com/code/rtatman/data-cleaning-challenge-handling-missing-values/input?select=NFL+Play+by+Play+2009-2016+%28v3%29.csvdf
df
```

```python
#Method - 3 - To analysis missing data

msno.heatmap(df) #as usual messy result due to so many columns are presented.
#will divide in 2 parts.
```

```python
df1 = df.iloc[:,:50]
msno.heatmap(df1) #df1 is contains same numbers of columns but only first 50 indexes(rows).

# Values close to positive 1 indicate that the presence of null values in one column 
# is correlated with the presence of null values in another column.

# Values close to negative 1 indicate that the presence of null values in one column 
# is anti-correlated with the presence of null values in another column.

# Values close to 0, indicate there is little to no relationship between the presence of null values 
# in one column compared to another.

#There are a number of values that show as <-1. This indicates that the correlation is very close to being 100% negative.
```

```python
df2 = df.iloc[:,50:]
msno.heatmap(df2) #df2 is contains same numbers of columns and remain columns after df1.
```
