```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Data-10x10-with_NaN_ascending_order.csv")
df
```

```python
#Method: 1 - linear

df.interpolate(method ='linear', limit_direction ='forward') 

# Great, when data points are in linear manner

# Linear Interpolation - The linear method ignores the index and treats missing values as equally spaced,
#and finds the best point to fit the missing value.

# limit_direction ='forward' - finds the missing value after previous points - THINK ABOUT TOP-DOWN APPROCH!
```

```python
df.interpolate(method ='linear', limit_direction ='backward')

# limit_direction ='backward' - Now the method will work from the end of the data frame - BOTTOM-UP APPROCH!
```

```python
# We can also do on only one column

df['A'].interpolate(method="linear") # By default it is forward and if we use just, ".interpolate()" then it will use linear in forward
```
