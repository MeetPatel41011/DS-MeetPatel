# Remove Duplications of Indexes (Rows)

```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
df #there are some duplicate Indexes (Rows)
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(keep='first') #'keep=first' means that the first occurrence of each duplicated row will be kept, and all subsequent duplicates will be removed.

#For example,

#df is,

# john 23 Canada
# wick 34 NZ
# john 23 Canada

#Result in "keep=first"

# john 23 Canada
# wick 34 NZ

#It keeps first occurance of "John 23 Canada" and removes second one.



#Additionally, as you can see above and below cells' outputs, both are same because "df.drop_duplicates()", use "keep='first'" as default
```

```python
df.drop_duplicates( keep='last') #'keep=last' means that the last occurrence of each duplicated row will be kept, and all subsequent duplicates will be removed.

# Vice versa - as "keep=first" - it keeps last one.
```

```python
df.drop_duplicates(keep=False) #Removes all duplicate indexes
```

```python
df.drop_duplicates(keep='first')
```

```python
df.drop_duplicates(keep='last')
```
